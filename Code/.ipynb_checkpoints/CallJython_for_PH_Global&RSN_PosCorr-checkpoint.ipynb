```python
t0 = time.time()
for RSN in RSNs7:
    t1 = time.time()
    indices = result_dict[RSN]
    print(RSN, len(indices))#, indices)
    rsn = RSN.replace(' ','')
    max_filtration_value = str(np.sqrt(2))
    max_dimension = str(2)
    
    command = ["java",
        # "-Djava.class.path=C:/jython2.7.3/jython.jar;C:/Users/hp/Desktop/IMSc/Projects/PersistentHomology_MPILEMON/MinScaffold-master_22Sep/Holes/jython_utils/lib/javaplex.jar",
        "-Djava.class.path=C:/jython2.7.3/jython.jar;C:/Users/hp/Desktop/IMSc/Projects/PersistentHomology_MPILEMON/Code_Javaplex/lib/javaplex.jar",
        "org.python.util.jython",
        code, dataset, path_file, RSN, RSNs_details_path, max_filtration_value, max_dimension]
    
    process = Popen(command, stdout=PIPE, stderr=PIPE, text=True)
    stdout, stderr = process.communicate()
    if stdout:
        print("Standard Output:\n", stdout)
    if stderr:
        print("Error Output:\n", stderr)
    
    if process.returncode != 0:
        print(f"Subprocess exited with error code {process.returncode}")
    else:
        print("Jython script executed successfully!", time.time()-t1, '-'*80)
    break
print('Done for',dataset, time.time()-t0, '-'*68)

print(path_file)
```

```python
Complex = {'vertex':[], 'edges':[]}
max_index = 1.4142
for key in Clique_dictionary:
    original_key=key
    key=str(key)  
    key=key.strip('[]')
    key=key.split(', ')
    
    key_buona=[]
    for n in range(len(key)):
        key_buona.append(int(float(eval(key[n]))))
    print(original_key, type(original_key), key,type(key),key_buona[0], type(key_buona[0]))
    # print(key_buona, Complex)
    if len(key_buona)==1:
        Complex['vertex'].append((key_buona[0],0))
    else:	
        Complex['edges'].append((key_buona, float(Clique_dictionary[original_key][1])))
        if float(Clique_dictionary[original_key][1])>max_index:
            max_index=int(Clique_dictionary[original_key][0])
    
```
